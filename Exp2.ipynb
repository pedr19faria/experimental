```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
f = lambda x, A, B, T, phi: A*np.sin(2*np.pi*x/T + phi) + B
y = f(x, 2.,  0.,5, 0.)
```

```python
def func(y):
    for i in range(len(y)):
        if y[i]< 0:
            y[i] = 0
    return y
y2 = func(y)
y = f(x, 2.,  0.,5, 0.)
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize = (10,4))
ax1.plot(x, y, label = "Posição A", c='orange', linewidth=2.5)
ax2.plot(x,y2, label= "Posição B", linewidth=2.5)
ax1.grid()
ax2.grid()
ax1.legend()
ax2.legend()
plt.show()
```

$$
V_{med} = \dfrac{V_{pp}}{2\pi}
$$

```python
vpp = 18
deltavpp = 1
Vmed = vpp/(2*np.pi)
incvpp = lambda vpp: np.sqrt( ((1/(2*np.pi) )**2)*((deltavpp)**2) )
deltavmed = incvpp(vpp)
print(f'Incerteza Vmed = ({Vmed:.1f} +- {deltavmed:.1f}) V')
```

```python
B = 0.4e-3
b = 0.2e-3
h=6
T =1e-3
volt = lambda B,b,h,T :((B+b)*h)/(2*T)
r = volt(B,b,h,T)
r
```

```python
incB = 0.05e-3
incb = 0.05e-3
inch = 1
incT = 0.05e-3

def incerteza(B,b,h,T, incB,incb, inch, incT):
    partialB = ((1)*h)/(2*T)
    partialb = ((1)*h)/(2*T)
    partialh = ((B+b))/(2*T)
    partialT = -(((B+b)*h)/(T**2))
    
    inc = np.sqrt( (partialB**2)*(incB**2)+(partialb**2)*(incb**2)
                 + (partialh**2)*(inch**2) +(partialT**2)*(incT**2))
    return inc

deltar = incerteza(B,b,h,T, incB,incb, inch, incT)

print(f'Incerteza V = ({r:.1f} +- {deltar:.1f}) V')
```
